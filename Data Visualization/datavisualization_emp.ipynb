##### **1.Salary distribution analysis**

```python
plt.figure(figsize=(10,6))
sns.histplot(emp_data['salary'],
            bins=20,kde=True,color='red'
            )
plt.title('Salary Distribution Across Employee',fontsize=14)
plt.xlabel('Salary',fontsize=12)
plt.ylabel('count',fontsize=12)
plt.grid(True,alpha = 0.3)
plt.show()
```

###### **Salary range is wide**-->Employee salaries span from low to high values, indicating multiple experience or role levels within the organization.

###### **Most employees fall in the mid-salary band**--> The KDE peak shows the highest concentration in the middle range, suggesting this is the core workforce segment.

###### **Distribution is right-skewed**-->A longer tail on the higher end indicates fewer high-paid employees compared to mid-level earners.

###### **Lower-salary group exists but is smaller**-->A small cluster at the lower end likely represents entry-level or junior employees.

#### **2.salary comparison based on promotion status**

```python
emp_melt_data=pd.melt(emp_data,
                     id_vars=['promotion_hr_based'],
                     value_vars=['salary','final_incremented_salary'],
                     var_name='Salary_Type',
                     value_name='Salary')
plt.figure(figsize=(10,6))
sns.barplot(x='promotion_hr_based',
           y='Salary',
           hue='Salary_Type',
           palette='Set2',
           data = emp_melt_data)
plt.title('Salary Promotion By Promotion Status',fontsize = 14)
plt.xlabel('promotion HR Data',fontsize = 12)
plt.ylabel('Salary',fontsize = 12)
plt.legend(title='Salary Type')
plt.show()
```

###### **Promotion pipeline employees have higher base salaries**-->Employees in the promotion pipeline earn more even before increment compared to those not in the pipeline

###### **Salary increases apply to all employees**-->Both promotion and non-promotion groups show an increase from salary  final incremented salary.

###### **Promotion leads to larger salary growth**-->The difference between original and incremented salary is larger for promotion-pipeline employees.

###### **Non-promotion employees receive smaller increments**-->Employees not in the promotion pipeline have lower absolute increments.

#### **3.Increment percentage evaluation**

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='promotion_hr_based',
           y='increment_percentage',
           hue='promotion_hr_based',
            palette='Set2',
           data=emp_data)
plt.title('Increament Percentage', fontsize = 14)
plt.xlabel('Promotion HR Based', fontsize =12)
plt.ylabel('Increament Percentage (%)', fontsize =12)
plt.legend(title='Promotion Status')
plt.show()
```

###### **Promotion pipeline employees receive higher increments**-->The median increment percentage for employees in the promotion pipeline is higher than for those not in the pipeline.

###### **Increment policy is more consistent for promoted employees**-->The narrow box (low IQR) for promotion-pipeline employees shows very little variation.

###### **Non-promotion employees show wide variability**-->The wider box and longer whiskers for non-promotion employees indicate large variation.



#### **4.Relationship between experience and salary growth**

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='experience_years',
               y='salary',
                hue='promotion_hr_based',
                size='increment_percentage',
                style='promotion_hr_based',
                palette='deep',
               sizes=(50,200),
               data=emp_data)
```

###### **Salary generally increases with experience**

###### **Promotion pipeline employees earn higher salaries at similar experience**

###### **Promotion pipeline employees receive higher increments**

###### **Non-promotion employees show wide salary variation**


#### **5.Sales Vs Productivity**

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='time_productive',
               y='sales',
               hue='experience_years',
               palette='viridis',
               size='experience_years',
               data=emp_data)
plt.title('Sales Vs Productivity', fontsize=14)
plt.xlabel('Productivity Time')
plt.ylabel('Sales', fontsize=12)
plt.show()
```

#### Insights:
###### **Sales show a positive relationship with productivity time.**
###### **At similar productivity levels, more experienced employees tend to generate higher sales.**
###### **Employees with around 23 years of experience appear to perform strongly in sales..**
###### **Sales outcomes vary at the same productivity level, suggesting experience or skill differences.**
###### **Some low-experience employees achieve high sales, indicating potential high performers..**

#### **Promotion hours vs exprience**

```python
promotion_data=emp_data.groupby('experience_years')['time_productive'].mean().reset_index()
plt.figure(figsize=(8,6))
sns.lineplot(data=promotion_data,
             x='experience_years',
            y='time_productive',
             palette='flare')
plt.title('Promotion Hours Vs Exprience')
plt.xlabel('Exprience Years')
plt.ylabel('Promotion(Productive Hours)')
plt.show()
```

#### Insights:
###### **Productive hours decrease between 1 and 2 years of experience.**
###### **Productive hours peak around 3 years of experience.**
###### **After around 3 years of experience, productive hours tend to decline.**
###### **More experienced employees appear to spend fewer hours on direct productive work..**
